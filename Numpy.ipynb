### wap to add the two matrices for a given number

```python
import numpy as np
a1=np.array([1,2,3,4,5])
a2=np.array([6,7,8,9,10])
a3=a1+a2
print(a3)
np.add(a1,a2)
np.matmul(a1,a2)
print(a1.dot(a2))
```

## indexing slicing and iterating

```python
import numpy as np
n1=np.arange(12).reshape(3,4)
print(n1)
print(" ")
print(n1[1:,1:3])
print(n1[2,:])
print(n1[2,3])
```

```python
import numpy as np
n1=np.arange(12).reshape(3,4)
print(n1)
#create n2 array out of n1 making first column as zero
n2=n1
n2[0,0]=0
n2[1,0]=0
n2[2,0]=0
print(n2)

```

```python
import numpy as np
n1=np.arange(12).reshape(3,4)
print(n1)
n2=n1
#n2[:,1:2]=0
n2[0,...]=1
print(n2)
n2[...,1]=2
print(n2)
n2[...,2]=8
print(n2)
```

```python
import numpy as np
a=np.arange(24).reshape(2,3,4)
print(a)
a1=a
a1[0,...,2]=0 # in these 0 is denoted first matrices in 3 dimensinal array
print(a1)
a1[1,...,2]=2# in these 1 is denoted second matrices in 3 dimensional array
print(a1)
```
